```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dataset/used_cars.csv')
df
```

```python
df.replace(['?', 'na', 'nil', ' '], np.nan, inplace=True)
df.isna().sum()
```

```python
df_cleaned = df.dropna(axis=0)
df_cleaned
```

```python
df_cleaned['milage'] = df_cleaned['milage'].str.replace(',', '').str.replace('mi', '').str.replace('.', '').astype(int)
```

```python
df_cleaned
```

```python
df_cleaned.milage
```

```python
df_cleaned['price'] = df_cleaned['price'].str.replace('$', '').str.replace(',', '').astype(int)
```

```python
df_cleaned.price
```

```python
df_cleaned.reset_index(inplace=True)
```
