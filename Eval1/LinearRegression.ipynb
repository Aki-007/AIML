```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('insurance.csv')
df
```

```python
df.replace(['?', 'nan', ' ', 'nil'], np.nan, inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
df['charges'] = round(df.charges, 2)
```

```python
df
```

```python
X = df.drop('charges', axis=1)
Y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one = OneHotEncoder(sparse_output=False, drop='first')
```

```python
x_train_encoded = one.fit_transform(x_train.select_dtypes(include='object'))
x_test_encoded = one.transform(x_test.select_dtypes(include='object'))
```

```python
x_train_encoded
```

```python
x_train_encoded = pd.DataFrame(x_train_encoded, columns=one.get_feature_names_out())
x_test_encoded = pd.DataFrame(x_test_encoded, columns=one.get_feature_names_out())
x_train_encoded
```

```python
x_train = pd.concat([x_train.select_dtypes(exclude='object').reset_index(), x_train_encoded], axis=1)
x_test = pd.concat([x_test.select_dtypes(exclude='object').reset_index(), x_test_encoded], axis=1)

x_train.drop('index', axis=1, inplace=True)
x_test.drop('index', axis=1, inplace=True)

x_train
```

```python
x_train.bmi.describe()
```

```python
y_train.dtype
```

```python
from sklearn.preprocessing import StandardScaler
y_scaler = StandardScaler()

y_train_scaled = y_scaler.fit_transform(y_train.values.reshape(-1, 1))
```

```python
y_train_scaled
```

```python
y_train = pd.DataFrame(y_train_scaled, columns=['charges'])
y_train
```

```python
y_test_scaled = y_scaler.transform(y_test.values.reshape(-1, 1))
```

```python
import torch
from torch import nn
from torch.nn import functional as f
```

```python
x_train_tensor = torch.tensor(x_train.values, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test_scaled, dtype=torch.float32)
```

```python
class LinearRegressionModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.linear = nn.Linear(in_features=8, out_features=1)

  def forward(self, x):
    return self.linear(x)
```

```python
model = LinearRegressionModel()
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(params=model.parameters(), lr=0.0001)
```

```python
#Training Loop
for i in range(100000):
  model.train()
  optimizer.zero_grad()
  output = model(x_train_tensor)
  loss = loss_fn(output, y_train_tensor)

  loss.backward()
  optimizer.step()

  if (i%10 == 0):
    print(f'Epoch[{i+1}/100000]', 'loss:', loss.item())
```

```python
model.eval()
with torch.inference_mode():
  test_output = model(x_test_tensor)
  test_loss = loss_fn(test_output, y_test_tensor)

print(f'Test Loss: {test_loss.item()}')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
y_pred_unscaled = y_scaler.inverse_transform(test_output.detach().numpy())
y_test_unscaled = y_scaler.inverse_transform(y_test_tensor.detach().numpy())

mae = mean_absolute_error(y_test_unscaled, y_pred_unscaled)
mse = mean_squared_error(y_test_unscaled, y_pred_unscaled)
r2 = r2_score(y_test_unscaled, y_pred_unscaled)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-squared (R2 Score): {r2:.2f}')
```
